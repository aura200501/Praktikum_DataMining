```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    port="3307"
)

cursor = conn.cursor()

# buat database
cursor.execute("CREATE DATABASE IF NOT EXISTS dbkampus")

# pilih database
cursor.execute("USE dbkampus")

# baca file SQL
with open("dbkampus.sql", "r") as file:
    sql_script = file.read()

# eksekusi
for statement in sql_script.split(";"):
    if statement.strip():
        cursor.execute(statement)

conn.commit()
print("Import berhasil!")
```

```python
# ubah id number jadikan integer karena akan membuang angka decimal di akhir
df_excel['ID number'] = df_excel ['ID number'].fillna(0).astype(int).astype(str)
df_excel.head()

#tambahan "0" didepan ID number : pola dari nim
df_excel['ID number'] = "0" + df_excel['ID number']
df_excel.head()
```

```python
# Standarisasi type data. agar aman samakan tipe data nya dahulu, dari excel ID number dari mysql nim

df_excel = df_excel.copy()
df_mysql = df_mysql.copy()

df_excel['ID number'] = df_excel['ID number'].astype(str)
df_mysql['nim'] = df_mysql['nim'].astype(str)
```

```python
# Integrasi (penggabungan data)

df_integrasi = pd.merge(
    df_mysql,df_excel,
    left_on='nim',
    right_on='ID number',
    how='inner'

)

df_integrasi.head()
```

```python
# buat dataframe agar lebih rapih
df_integrasi = df_integrasi[['id', 'nim', 'nama','Grade/100.00']]

df_integrasi.head()
```

```python
# rename Grade/100.00 menjadi nilai

df_integrasi.rename(columns={'Grade/100.00': 'nilai'}, inplace=True)

df_integrasi.head()
```
